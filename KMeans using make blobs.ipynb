```python
#Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
```

```python
#Generate dataset
X,y = make_blobs(n_samples = 100, centers = 3, cluster_std = 1, shuffle = True)
```

```python
#Visualise the data
plt.scatter(X[:,0], X[:,1])
```

```python
#Establish the right number of clusters
#Generate and empty list and create a range that will go from 1 to 8
empty = []
r = range(1,9)

#Establish the inertia for the number of clusters between 1 to 8 and add this to list 'empty'
for i in r:
    km = KMeans(n_clusters = i)
    km.fit_predict(X)
    empty.append(km.inertia_)
  
#Establish the inflection point to assess the right number of clusters
plt.plot(r,empty)
```

```python
#Rerun kmeans with 3 clusters
km2 = KMeans(n_clusters = 3)
y = km2.fit_predict(X)
```

```python
#Confirm that the labels allocated look sensible on the visualisation 
#Plot the datapoints labelled as 0 as red
plt.scatter(X[y == 0,0],X[y==0,1], c = 'r')

#Plot the datapoints labelled as 1 as blue
plt.scatter(X[y == 1,0],X[y==1,1], c = 'b')

#Plot the datapoints labelled as 2 as green
plt.scatter(X[y == 2,0],X[y==2,1], c = 'g')

plt.show()
```

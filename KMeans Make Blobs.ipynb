```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
import seaborn as sns
```

```python
X, y = make_blobs(n_samples = 150, centers = 3, shuffle = True, random_state = True)
plt.scatter(x = X[:,0], y = X[:,1])
```

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters = 3, init = 'random')
y_km = model.fit_predict(X)
model.cluster_centers_
```

```python
plt.scatter(
    X[y_km == 0,0],
    y = X[y_km == 0,1],
    color = 'k',
    label = 'cluster 1',
    marker = 's'
)

plt.scatter(
    X[y_km == 1,0],
    X[y_km == 1,1],
    color = 'r',
    label = 'cluster 2',
    marker = 'o')


plt.scatter(
    X[y_km == 2,0],
    X[y_km == 2,1],
    label = 'cluster 3',
    marker = 'x')


plt.scatter(
   model.cluster_centers_[:,0],
    model.cluster_centers_[:,1],
    s = 100,
    color = 'y',
    label = 'centroids'
      )

plt.legend()

plt.show()
```
